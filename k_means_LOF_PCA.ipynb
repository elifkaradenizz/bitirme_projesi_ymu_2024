# 1) RFM


RFM, "Recency, Frequency, Monetary" kelimelerinin ba harflerinden oluan bir mteri segmentasyon yntemidir. Bu yntem, mterileri davranlarna gre segmentlere ayrmak ve bu segmentlere zg pazarlama stratejileri oluturmak iin kullanlr. RFM analizi, her mterinin  ana kriter olan Recency (Yenilik), Frequency (Sklk) ve Monetary (Parasal Deer) zelliklerine dayanr.

Recency (Yenilik): Mterinin son satn almasndan bu yana geen sredir. Bu kriterde daha yakn zamanda satn alan mteriler daha yksek bir puan alr.

Frequency (Sklk): Belirli bir zaman diliminde mterinin yapt toplam satn alma saysdr. Sklk arttka mterinin puan da artar.

Monetary (Parasal Deer): Mterinin toplam harcad para miktardr. Daha yksek miktarlarda harcama yapan mteriler daha yksek bir puan alr.

A) Recency(R)

Uygulayacam mantk --> Son Satn Almadan Bu Yana Geen Gnler: Bu zellik, mterinin son satn alma ileminden bu yana geen gn saysn temsil eder. Daha dk bir deer, mterinin yakn zamanda alveri yaptn ve iletmeyle daha yksek bir ballk dzeyine sahip olduunu gsterirken, daha yksek bir deer bir duraklamaya veya azalan balla iaret edebilir.

```python
# InvoiceDate'i datetime trne dntr. Bu ilem, satn alma tarihlerinin uygun formatta saklanmasn salar.
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# InvoiceDate'i datetime'a dntrn ve yalnzca tarihi karn.
df['InvoiceDay'] = df['InvoiceDate'].dt.date

# Her mteri iin en son satn alma tarihini bulun
customer_data = df.groupby('CustomerID')['InvoiceDay'].max().reset_index()

"""Tm veri kmesindeki en son tarihi buluyoruz. Bu sayede en son tarihten her mterinin en son satn alma tarihini karcaz.Daha dk bir deer, mterinin yakn zamanda alveri
yaptn daha byk deer ise uzun zaman nce alveri yaptn gstericek. """

most_recent_date = df['InvoiceDay'].max()

# karma ileminden nce InvoiceDay'i datetime trne dntrn
customer_data['InvoiceDay']= pd.to_datetime(customer_data['InvoiceDay'])
most_recent_date = pd.to_datetime(most_recent_date)

# Her mteri iin son satn alma ileminden bu yana geen gn saysn hesaplayn
customer_data['Days_Since_Last_Purchase'] = (most_recent_date - customer_data['InvoiceDay']).dt.days

# InvoiceDay stununu kaldrn
customer_data.drop(columns=['InvoiceDay'], inplace=True)
```

b) Frequency (F)

Uygulayacam mantk --> Bu admda, bir mterinin perakendeciyle etkileim skln len iki zellik oluturacam:

Toplam lemler: Bu zellik, bir mteri tarafndan yaplan toplam ilem saysn temsil eder. Bir mterinin perakendeciyle olan etkileim dzeyini anlamaya yardmc olur.

Satn Alnan Toplam rn: Bu zellik, bir mteri tarafndan tm ilemlerde satn alnan toplam rn saysn (miktarlarn toplam) gsterir. Satn alnan rnlerin hacmi asndan mterinin satn alma davran hakknda bir fikir verir.

```python
# Her bir mteri tarafndan yaplan toplam ilem saysn hesaplayn
total_transactions = df.groupby('CustomerID')['InvoiceNo'].nunique().reset_index()
total_transactions.rename(columns={'InvoiceNo': 'Total_Transactions'}, inplace=True)

# Her bir mteri tarafndan satn alnan toplam rn saysn hesaplayn
total_products_purchased = df.groupby('CustomerID')['Quantity'].sum().reset_index()
total_products_purchased.rename(columns={'Quantity': 'Total_Products_Purchased'}, inplace=True)

# Yeni zellikleri customer_data veri erevesiyle birletirin
customer_data = pd.merge(customer_data, total_transactions, on='CustomerID')
customer_data = pd.merge(customer_data, total_products_purchased, on='CustomerID')
```

c) Monetary (M)

Bu admda mteri ilemlerinin parasal ynn temsil eden iki zellik oluturacam:

Toplam Harcama: Bu zellik, her mterinin harcad toplam para miktarn temsil eder. Mterinin yapt tm ilemler iin BirimFiyat ve Adet arpmnn toplam olarak hesaplanr. Bu zellik, mterinin iletme iin deerinin dorudan bir gstergesi olan, her mterinin rettii toplam gelirin belirlenmesine yardmc olduu iin ok nemlidir.
Ortalama lem Deeri: Bu zellik, her mteri iin Toplam Harcamann Toplam lemlere blnmesiyle hesaplanr. Mteri tarafndan gerekletirilen bir ilemin ortalama deerini gsterir. Bu lm, mterilerin ilem bana harcama davranlarn anlamada kullanldr ve ortalama harcama kalplarna gre farkl mteri segmentlerine ynelik pazarlama stratejileri ve tekliflerin uyarlanmasna yardmc olabilir.

```python
# Her bir mterinin toplam harcamasn hesaplayn
df['Total_Spend'] = df['UnitPrice'] * df['Quantity'] #rn fiyat x satn alnan miktar
total_spend = df.groupby('CustomerID')['Total_Spend'].sum().reset_index()

# Her mteri iin ortalama ilem deerini hesaplayn
average_transaction_value = total_spend.merge(total_transactions, on='CustomerID')# total_spend veri erevesi ile total_transactions veri erevesini CustomerID
# stunu zerinden birletiriyoruz (merge() fonksiyonu ile). Bu sayede her mterinin toplam harcamasn (Total_Spend) ve toplam ilem saysn (Total_Transactions) ieren bir veri erevesi elde ediyoruz.
average_transaction_value['Average_Transaction_Value'] = average_transaction_value['Total_Spend'] / average_transaction_value['Total_Transactions'] # Daha sonra bu birletirilmi veri erevesi zerinden her mteri iin ortalama ilem deerini (Average_Transaction_Value) hesaplyoruz.
# Ortalama ilem deeri, mterinin toplam harcamas (Total_Spend) ile yapt ilem says (Total_Transactions) arasndaki blme ilemiyle hesaplanyor (Total_Spend / Total_Transactions).

# Yeni zellikleri customer_data veri erevesiyle birletirin
customer_data = pd.merge(customer_data, total_spend, on='CustomerID')
customer_data = pd.merge(customer_data, average_transaction_value[['CustomerID', 'Average_Transaction_Value']], on='CustomerID')
```

# 2) Satn Alnan rn eitliliini nceleme

# 6) Trendleri Ele Alma


Mterilerin harcama trendlerini analiz etmek, alveri davranlarnn zaman iindeki deiimlerini belirleyerek pazarlama stratejilerini ayarlamak ve mteri memnuniyetini artrmak iin nemlidir.

```python
from scipy.stats import linregress

# InvoiceDate'ten ay ve yl kartn
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month

# Her mteri iin aylk harcamay hesaplayn
monthly_spending = df.groupby(['CustomerID', 'Year', 'Month'])['Total_Spend'].sum().reset_index()

# Sezonluk Satn Alma Modellerini Hesaplayn: Sezonluk satn alma modellerinin temsili olarak aylk skl kullanlyor..
seasonal_buying_patterns = monthly_spending.groupby('CustomerID')['Total_Spend'].agg(['mean', 'std']).reset_index()
seasonal_buying_patterns.rename(columns={'mean': 'Monthly_Spending_Mean', 'std': 'Monthly_Spending_Std'}, inplace=True)

# Monthly_Spending_Std'deki NaN deerlerini 0 ile deitirin; bu, tek ilem ayna sahip mteriler iin hibir deikenlik olmad anlamna gelir..
seasonal_buying_patterns['Monthly_Spending_Std'].fillna(0, inplace=True)

# Harcama Eilimlerini Hesaplayn. Harcama eilimlerinin bir gstergesi olarak mterinin zaman iindeki harcamalarna uygun dorusal eilim izgisinin eimini kullanyoruz..
def calculate_trend(spend_data):
    # Birden fazla veri noktas varsa trendi dorusal regresyon kullanarak hesaplarz
    if len(spend_data) > 1:
        x = np.arange(len(spend_data))
        slope, _, _, _, _ = linregress(x, spend_data)
        return slope
  # Yalnzca tek bir veri noktas varsa hibir eilim hesaplanamaz, dolaysyla 0 dndrrz
    else:
        return 0

# Her mterinin harcama eilimini bulmak iin spending_trends ilevini uygulayn
spending_trends = monthly_spending.groupby('CustomerID')['Total_Spend'].apply(calculate_trend).reset_index()
spending_trends.rename(columns={'Total_Spend': 'Spending_Trend'}, inplace=True)

# Yeni zellikleri customer_data veri erevesiyle birletirin
customer_data = pd.merge(customer_data, seasonal_buying_patterns, on='CustomerID')
customer_data = pd.merge(customer_data, spending_trends, on='CustomerID')

customer_data.head()
```

# **K-MEANS CLUSTERING START**

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Elbow yntemiyle uygun kme saysn belirleme ilevi
def calculate_WSS(points, kmax):
    sse = []
    for k in range(1, kmax + 1):
        kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)  # n_init parametresi belirtilerek kullanlyor
        kmeans.fit(points)
        sse.append(kmeans.inertia_)  # k-means modelinin inertia deerini WSS skoru olarak kaydeder
    return sse

# PCA dnml veri setini kullanarak WSS skorlarn hesapla
kmax = 15  # Denenecek maksimum kme says
wss_scores = calculate_WSS(df_pca, kmax)

# Elbow grafii izdirme
plt.figure(figsize=(8, 6))
plt.plot(range(1, kmax + 1), wss_scores, marker='o', linestyle='--')
plt.xlabel('Kme Says (k)')
plt.ylabel('Within-Cluster Sum of Squares (WSS) Skoru')
plt.title('Elbow Yntemi ile Optimal Kme Saysnn Belirlenmesi')
plt.xticks(range(1, kmax + 1))
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import matplotlib.pyplot as plt

# Kme says iin denenecek aralk
min_k = 2
max_k = 10

# Metrik deerlerini depolamak iin bo listeler oluturun
sil_scores = []
ch_scores = []
db_scores = []

# Her bir kme says iin metrikleri hesaplayn
for k in range(min_k, max_k + 1):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(df_pca)
    labels = kmeans.labels_

    # Siluet skoru hesapla
    sil_avg = silhouette_score(df_pca, labels)
    sil_scores.append(sil_avg)

    # Calinski-Harabasz skoru hesapla
    ch_score = calinski_harabasz_score(df_pca, labels)
    ch_scores.append(ch_score)

    # Davies-Bouldin skoru hesapla
    db_score = davies_bouldin_score(df_pca, labels)
    db_scores.append(db_score)

# Metrik skorlarn grselletirme
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(range(min_k, max_k + 1), sil_scores, marker='o', linestyle='--', label='Silhouette Score')
plt.xlabel('Kme Says (k)')
plt.ylabel('Silhouette Skoru')
plt.title('Farkl Kme Saylar iin Performans Metrikleri')
plt.xticks(range(min_k, max_k + 1))
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(range(min_k, max_k + 1), ch_scores, marker='o', linestyle='--', color='orange', label='Calinski-Harabasz Score')
plt.xlabel('Kme Says (k)')
plt.ylabel('CH Skoru')
plt.xticks(range(min_k, max_k + 1))
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(range(min_k, max_k + 1), db_scores, marker='o', linestyle='--', color='green', label='Davies-Bouldin Score')
plt.xlabel('Kme Says (k)')
plt.ylabel('DB Skoru')
plt.xticks(range(min_k, max_k + 1))
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
